```python
def recursum(n):
  if n<=1:
    return n
  return n +recursum(n-1)
n=10
print(recursum(n))
```

```python
def recur_fib(n):
  if n<=1:
    return(n)
  else:
    return(recur_fib(n-1)+recur_fib(n-2))
n=5
for i in range(n):
  print(recur_fib(i))
```

```python
def recur_fact(n):
  if n==0:
    return 1
  else:
    return n*recur_fact(n-1)
n=4
print(recur_fact(n))
```

```python
#calculate the numbser of upper and lower case alphabets in a given string
#String : She sells Seashells by the Seashore
def calc_string(x):
  u_count=0
  l_count=0
  for char in x:
    if char.isupper():
      u_count+=1
    elif char.islower():
      l_count+=1
  return u_count, l_count
print(calc_string("She sells Seashells by the Seashore"))
```

```python
def cal_string(x):
  lower=0
  upper=0
  for char in x:
    if char.isupper()==True:
      upper+=1
    elif char.islower()==True:
      lower+=1
    return lower, upper
print(cal_string("She sells Seashells by the Seashore"))
```
